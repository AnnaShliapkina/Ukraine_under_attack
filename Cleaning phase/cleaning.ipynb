```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import os
import folium
```

```python
data_path='../data'
sns.set(rc={'figure.figsize':(20,6)})
```

```python
conflict_df=pd.read_csv(os.path.join(data_path,'2022-02-24-2024-04-20-Europe-Ukraine.csv'))
conflict_df.head()
```

```python
conflict_df.info()
```

```python
conflict_df.describe(include='all')
```

```python
# Створення нового датафрейму з копією конфліктного датафрейму та зміненим стовпцем "event_id_cnty"
expl_df = conflict_df.copy()




# Виведення нового датафрейму
expl_df.head(2)
```

```python
# Вибираємо тільки потрібні стовпці та фільтруємо за потрібними значеннями у стовпці "event_type"
expl_df = conflict_df.loc[conflict_df['event_type'].isin(['Explosions/Remote violence', 'Riots', 'Violence against civilians']), 
                          ['event_id_cnty', 'event_date', 'year', 'event_type', 'sub_event_type', 'actor1', 'actor2', 'inter2', 
                           'civilian_targeting', 'admin1', 'admin2', 'location', 'latitude', 'longitude', 'source', 
                           'fatalities', 'population_best']]

# Перейменовуємо стовпці
expl_df.rename(columns={'event_id_cnty': 'event_id', 'admin1': 'region', 'admin2': 'district', 'population_best': 'population'}, inplace=True)

# Виділення останніх 6 символів та перетворення на ціле число
expl_df['event_id'] = expl_df['event_id'].str[3:].astype(int)

expl_df.reset_index(drop=True, inplace=True)

# Виводимо новий датафрейм
expl_df.head()
```

```python
# Заміна значень у колонці 'civilian_targeting' згідно з умовою
expl_df['civilian_targeting'] = expl_df['civilian_targeting'].apply(lambda x: 'Yes' if x == 'Civilian targeting' else 'No' if pd.isna(x) else x)

# Виведення оновленого датафрейму
expl_df.head()
expl_df['sub_event_type'].unique()
```
